# Hg38 mapping

```r
alignResult = c()
for(hist in all.sample$V1){
  # if(hist=="M60_K27Ac_1") break
  alignRes = read.table(paste0("bam/", hist, ".Map2GenomeStat.xls"), header = FALSE, fill = TRUE)
  alignRate = substr(alignRes$V1[6], 1, nchar(as.character(alignRes$V1[6]))-1)
  histInfo = strsplit(hist, "_")[[1]]
  Histone = paste(histInfo[1], histInfo[2], sep="_")
  Replicate = histInfo[3]
  alignResult = data.frame(Histone = Histone, Replicate = Replicate, 
                           SequencingDepth = alignRes$V1[1] %>% as.character %>% as.numeric, 
                           MappedFragNum_hg38 = alignRes$V1[4] %>% as.character %>% as.numeric + alignRes$V1[5] %>% as.character %>% as.numeric, 
                           AlignmentRate_hg38 = alignRate %>% as.numeric)  %>% rbind(alignResult, .)
}

# alignResult$Histone = factor(alignResult$Histone, levels = all.sample$V1)
# alignResult %>% mutate(AlignmentRate_hg38 = paste0(AlignmentRate_hg38, "%"))
rownames(alignResult) <- all.sample$V1
```

# E.coli mapping

```r
spikeAlign = c()
for(hist in all.sample$V1){
  spikeRes = read.table(paste0("Ecoli_bam/", hist, ".Ecoli.Map2GenomeStat.xls"), header = FALSE, fill = TRUE)
  alignRate = substr(spikeRes$V1[6], 1, nchar(as.character(spikeRes$V1[6]))-1)
  histInfo = strsplit(hist, "_")[[1]]
  Histone = paste(histInfo[1], histInfo[2], sep="_")
  Replicate = histInfo[3]
  spikeAlign = data.frame(Histone = Histone, Replicate = Replicate,
                          SequencingDepth = spikeRes$V1[1] %>% as.character %>% as.numeric, 
                          MappedFragNum_spikeIn = spikeRes$V1[4] %>% as.character %>% as.numeric + spikeRes$V1[5] %>% as.character %>% as.numeric, 
                          AlignmentRate_spikeIn = alignRate %>% as.numeric)  %>% rbind(spikeAlign, .)
}
# spikeAlign$Histone = factor(spikeAlign$Histone, levels = histList)
# spikeAlign %>% mutate(AlignmentRate_spikeIn = paste0(AlignmentRate_spikeIn, "%"))
rownames(spikeAlign) <- all.sample$V1
```

# full QC info
- Sequencing depth
- Alignment rate
- Number of mappable fragments
- Duplication rate
- Unique library size
- Fragment size distribution

- the percentage of E. coli reads range from ~0.01% to 10%. 
- With fewer cells or less abundant epitopes, E. coli reads can comprise as much as 70% or the total mapped reads. 
- For IgG controls, the percentage of E. coli reads is typically much higher than that for an abundant histone modification.

## duplication

```r
##=== R command ===## 
## Summarize the duplication information from the picard summary outputs.
dupResult = c()
for(hist in all.sample$V1){
  dupRes = read.table(paste0("bam/", hist, "_picard.rmDup.txt"), header = TRUE, fill = TRUE)
  
  histInfo = strsplit(hist, "_")[[1]]
  Histone = paste(histInfo[1], histInfo[2], sep="_")
  Replicate = histInfo[3]
  dupResult = data.frame(Histone = Histone, Replicate = Replicate, 
                         MappedFragNum_hg38 = dupRes$READ_PAIRS_EXAMINED[1] %>% as.character %>% as.numeric, DuplicationRate = dupRes$PERCENT_DUPLICATION[1] %>% as.character %>% as.numeric * 100, EstimatedLibrarySize = dupRes$ESTIMATED_LIBRARY_SIZE[1] %>% as.character %>% as.numeric) %>% mutate(UniqueFragNum = MappedFragNum_hg38 * (1-DuplicationRate/100))  %>% rbind(dupResult, .)
}
# dupResult$Histone = factor(dupResult$Histone, levels = histList)
alignDupSummary = left_join(alignSummary, dupResult, by = c("Histone", "Replicate", "MappedFragNum_hg38")) %>% mutate(DuplicationRate = paste0(DuplicationRate, "%"))
alignDupSummary
```

```r
##=== R command ===## 
## generate boxplot figure for the  duplication rate
fig4A = dupResult %>% ggplot(aes(x = Histone, y = DuplicationRate, fill = Histone)) +
    geom_boxplot() +
    geom_jitter(aes(color = Replicate), position = position_jitter(0.15)) +
    scale_fill_viridis(discrete = TRUE, begin = 0.1, end = 0.9, option = "magma", alpha = 0.8) +
    scale_color_viridis(discrete = TRUE, begin = 0.1, end = 0.9) +
    theme_bw(base_size = 18) +
    ylab("Duplication Rate (*100%)") +
    xlab("") 

fig4B = dupResult %>% ggplot(aes(x = Histone, y = EstimatedLibrarySize, fill = Histone)) +
    geom_boxplot() +
    geom_jitter(aes(color = Replicate), position = position_jitter(0.15)) +
    scale_fill_viridis(discrete = TRUE, begin = 0.1, end = 0.9, option = "magma", alpha = 0.8) +
    scale_color_viridis(discrete = TRUE, begin = 0.1, end = 0.9) +
    theme_bw(base_size = 18) +
    ylab("Estimated Library Size") +
    xlab("") 

fig4C = dupResult %>% ggplot(aes(x = Histone, y = UniqueFragNum, fill = Histone)) +
    geom_boxplot() +
    geom_jitter(aes(color = Replicate), position = position_jitter(0.15)) +
    scale_fill_viridis(discrete = TRUE, begin = 0.1, end = 0.9, option = "magma", alpha = 0.8) +
    scale_color_viridis(discrete = TRUE, begin = 0.1, end = 0.9) +
    theme_bw(base_size = 18) +
    ylab("# of Unique Fragments") +
    xlab("")

ggarrange(fig4A, fig4B, fig4C, ncol = 3, common.legend = TRUE, legend="bottom")
```

# Spike-in calibration

```r
##=== R command ===## 
scaleFactor = c()
multiplier = 10000
for(hist in all.sample$V1){
  spikeDepth = read.table(paste0("Ecoli_bam/", hist, "_spikeIn.seqDepth"), header = FALSE, fill = TRUE)$V1[1]
  
  histInfo = strsplit(hist, "_")[[1]]
  Histone = paste(histInfo[1], histInfo[2], sep="_")
  Replicate = histInfo[3]
  scaleFactor = data.frame(scaleFactor = multiplier/spikeDepth, Histone = Histone, Replicate = Replicate)  %>% 
                    rbind(scaleFactor, .)
}
# scaleFactor$Histone = factor(scaleFactor$Histone, levels = histList)
qc_summary <- left_join(alignDupSummary, scaleFactor, by = c("Histone", "Replicate"))
```

```r
plot_heatmap_deeptools <- function(bigwig.files, samplesLabel, regions.file, out.dir, out.file.name,
                                   geneset.color="#F8766D", flank.len = 1000) {
    cmd1 <- sprintf("/home/zz950/softwares/miniconda3/bin/computeMatrix reference-point -S %s -R %s --referencePoint center -b %s -a %s -bs=1 -p=max --outFileName %s/%s.matrix.gz --missingDataAsZero", 
                    bigwig.files, regions.file, flank.len, flank.len, out.dir, out.file.name)
    cmd2 <- sprintf("/home/zz950/softwares/miniconda3/bin/plotHeatmap -m  %s/%s.matrix.gz -o %s/%s.heatmap.png --legendLocation best --colorList \"white,%s\" --samplesLabel %s", 
                    out.dir, out.file.name, out.dir, out.file.name, geneset.color, samplesLabel)
    system(cmd1, intern = T)
    system(cmd2, intern = T)
    # message(cmd1)
    # message(cmd2)
    message("Done!")
}
```

- must add `--missingDataAsZero`, otherwise they will be NaN and is black on background

```r
plot_heatmap_deeptools(bigwig.files, samplesLabel, regions.file, out.dir, out.file.name)
```
